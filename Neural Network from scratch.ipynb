### Importing the library

```python
import numpy as np
```

### Creatig the input for neural network

```python
input_feature = np.array([[0,0],[0,1],[1,0],[1,1]])
print(input_feature.shape)
input_feature
```

### Creating the output for neural network

```python
target_output = np.array([[0,1,1,1]])
print(target_output)
target_output = target_output.reshape(4,1)
print(target_output.shape)
target_output
```

### Creating the weights metric

```python
weights = np.array([[0.1],[0.2]])
weights.shape
```

### Creating bias and learning rate varibales

```python
bias = 0.3
lr = 0.05
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def sigmoid_der(x):
    return sigmoid(x) * (1-sigmoid(x))
```

### Constructing the neural network

```python
for epoch in range(1000):
    inputs = input_feature
    
    #forward propogation
    in_o = np.dot(inputs,weights) + bias
    
    out_o = sigmoid(in_o)
    
    #back propogation
    error = out_o-target_output
    
    x = error.sum()
    print(x)
    
    derror_douto = error
    douto_dino = sigmoid_der(out_o)
    deriv = derror_douto * douto_dino
#     print(deriv.shape)
    
    inputs = input_feature.T
    deriv_final = np.dot(inputs,deriv)
    
    weights -= lr*deriv_final
#     print(weights)
    
    for i in deriv:
        bias -= lr*i
#         print(bias)
```
